## Pivot Table

```python
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.groupby("sex")[["survived"]].mean()
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean").unstack() #unstack okumayı kolaylaştırır
```

```python
#pivot table
titanic.pivot_table("survived",index="sex",columns="class")
```

```python
titanic.age.head()
```

```python
age = pd.cut(titanic["age"],[0,18,90])
age.head(10)
```

```python
titanic.pivot_table("survived",["sex",age],"class") #dışarıdan alınan değişkenler ""sız yazılır.
```
